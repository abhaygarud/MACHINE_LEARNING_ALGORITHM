```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("full.csv")
df.head(10)
```

```python
df.drop(['Body','Lifeboat','Class','Name','SibSp','Parch','Ticket','Cabin','Embarked','Fare','WikiId','Name_wiki','Age_wiki','Hometown','Boarded','Destination'],axis='columns',inplace=True)
df   
```

```python
df.Survived=df.Survived.fillna(df.Survived.mean())
df.head(10)
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head(5)
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.drop(['Sex'],axis='columns')
```

```python
inputs.columns[inputs.isna().any()]
```

```python
target
```

```python
inputs.drop(['Sex','male'],axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.Age[:10]
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
X_train
```

```python
y_train
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
