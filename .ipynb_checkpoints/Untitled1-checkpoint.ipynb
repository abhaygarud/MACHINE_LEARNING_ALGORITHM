```python
import matplotlib.pyplot as plt
import numpy as np

Days=np.array([1,2,3,4,5,6,7])
max_t=np.array([50,55,53,56,66,45,51])
min_t=np.array([10,15,20,16,17,9,8])
avg_t=np.array([20,25,21,28,22,29,32])

plt.xlabel("days")
#y-axis label
plt.ylabel("temperature")
#title
plt.title("weather")
#ms is marker size ,mfc marker border color, alpha is colour
plt.plot(max_t,label="max",marker="D",ms=10,mfc="r",alpha=0.5)
plt.plot(min_t,label="mim")
plt.plot(avg_t,label="avg")
#legend is basicaly a scale
plt.legend(loc="best",shadow="avg",fontsize="small")
#grid is a line in a plane
plt.grid(color='green',linestyle='--',linewidth=0.5,alpha=0.5)
```
